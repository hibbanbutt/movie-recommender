```python
import pandas as pd

movie_processed = pd.read_parquet('/content/movie_processed_cleanedd.parquet')

movie_processed.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import umap
from sklearn.cluster import DBSCAN



# Apply TF-IDF on the `cleaned_description` column
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(movie_processed['cleaned_description'])

#  Apply UMAP transformations on the TF-IDF matrix
reducer = umap.UMAP()
umap_embeddings = reducer.fit_transform(tfidf_matrix)

# Apply DBSCAN clustering on the UMAP transformed data
dbscan = DBSCAN(eps=0.3, min_samples=3)
dbscan_labels = dbscan.fit_predict(umap_embeddings)

# Attach the cluster labels to the original dataframe
movie_processed['cluster_label'] = dbscan_labels
```
