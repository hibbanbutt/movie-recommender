```python
def update_description(row):
    if pd.isna(row["description"]) and row["title"] in manual_descriptions:
        return manual_descriptions[row["title"]]
    else:
        return row["description"]

movies_df["description"] = movies_df.apply(update_description, axis=1)
```

```python
import spacy

nlp = spacy.load("en_core_web_lg")

def get_document_vector(text):
    if text:
        doc = nlp(text)
        return doc.vector
    else:
        return None
def get_document_vector(text):
    if text:
        doc = nlp(text)
        return doc.vector
    else:
        return None
```

```python
movies_df["intro_vector"] = movies_df["description"].apply(get_document_vector)
```
