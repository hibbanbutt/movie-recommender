```python
import numpy as np
import pandas as pd
from sklearn.cluster import  AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score, fowlkes_mallows_score, mutual_info_score
from sklearn.preprocessing import LabelEncoder


# Read the parquet file
#df = pd.read_parquet('..\movies_processed.parquet')



movies_df = pd.read_parquet("..\movies_processed.parquet")
#Just in case: Drop any rows with missing intro_vector values:
movies_df = movies_df.dropna(subset=["intro_vector"])


# Extract the genre labels
true_labels = movies_df['genres'].values.tolist()
single_genre_labels = [label.split('|')[0] for label in true_labels]

#Create a NumPy array of intro_vectors for clustering:
intro_vectors = np.stack(movies_df["intro_vector"].values)
print(single_genre_labels)
```

```python
agg_clustering = AgglomerativeClustering(n_clusters=50, linkage='average')
cluster_labels = agg_clustering.fit_predict(intro_vectors)
```

```python
encoder = LabelEncoder()
true_labels_encoded = encoder.fit_transform(single_genre_labels)
```

```python
print(len(single_genre_labels))
print(len(intro_vectors))
```

```python
ari = adjusted_rand_score(true_labels_encoded, cluster_labels)
fmi = fowlkes_mallows_score(true_labels_encoded, cluster_labels)
mi = mutual_info_score(true_labels_encoded, cluster_labels)

print(f"Adjusted Rand Index: {ari}")
print(f"Fowlkes-Mallows score: {fmi}")
print(f"Mutual Information: {mi}")
```
